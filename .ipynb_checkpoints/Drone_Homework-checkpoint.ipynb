```python
# First we need to fetch some data using Python requests from API
# Read more about Python requests:
# http://docs.python-requests.org/en/master/user/quickstart/

import requests
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime

response = requests.get("http://api.dronestre.am/data")
json_data = response.json()
df = pd.DataFrame(json_data['strike'])
```

We wanted to concentrate first on the 'deaths' column as it struck us as the most important data point, so we printed it below to examen and found that there were issues with the data (not integers, some contained ranges with dashes, there were 'Unknown' (strings))

```python
df.deaths
```

The first step was to create a new DataFrame (df_deaths) that was the same as the origninal except it removed rows in which the 'deaths' column contained 'Unknown' values.  We thought about replacing the 'Unkown' values with the median value of the whole 'deaths' dataset, but that would skew the results (decrease the dispersion) in an artifial manner

```python
df_deaths = df[df['deaths'] != 'Unknown']
```

The next step is to write a lambda function to split the rows in the 'deaths' column by their dash creating a list of two separate values for those cells in which that is true

```python
df_deaths['avg_death'] = df_deaths['deaths'].apply(lambda x: x.split('-'))

df.head()
```

The cell below makes a new column which contains a single integer: 1) either the standalone integer in the 'death' column, or 2) the mean of the two numbers for the rows that have dashes.  The numbers are also then converted into integers

```python
df_deaths['avg_death'] = df_deaths['deaths'].apply(lambda x: x.split('-'))

death_list = []

for x in df_deaths['avg_death']:
    
    length = len(x)
    
    if length < 2:
        death_list.append(int(x[0]))
    
    else:
        first = int(x[0])
        second = int(x[1])
        avg = (first + second)/2
        death_list.append(avg)
        
df_deaths['avg_death'] = death_list

df_deaths
```

The goal below was to drop a couple of the huge outliers (a few points above 50) and charts a scatterplot to assess if the strikes have become deadlier (higher # of deaths) over time.  The result is that there is not much of a relationship over time (slight negative relationship)

```python
df_no_outliers = df_deaths[df_deaths['avg_death'] < 50] 

df_no_outliers.plot(kind='scatter', x='number', y='avg_death')
```

```python
from mpl_toolkits.basemap import Basemap
```

The below imports a library we downloaded called Basemap that allows you to chart using maps. It then creates a DataFrame with only the 'latitidue and longitude' columns in the original DataFrame.  It then creates two lists and function which turns the longitiude and latitude datapoints into floats and appends the two lists

```python
terror_data = df[['lat', 'lon']]

terror_data = terror_data[terror_data['lon'] != '']


latitude = []
longitude = []

def to_float(value, lst):
    for x in value:
        lst.append(float(x))
    return lst

latitude2 = to_float(terror_data['lat'], latitude)
longitude2 = to_float(terror_data['lon'], longitude)
```

The below creates blue dots on where the strikes occur on the World Map

```python
plt.figure(1, figsize=(12,6))

m1 = Basemap(projection='merc',
             llcrnrlat=-5,
             urcrnrlat=45,
             llcrnrlon=15,
             urcrnrlon=100,
             lat_ts=0,
             resolution='c')

m1.fillcontinents(color='#191919',lake_color='#000000')
m1.drawmapboundary(fill_color='#000000')
m1.drawcountries(linewidth=0.1, color="w")

mxy = m1(longitude2, latitude2)
m1.scatter(mxy[0], mxy[1], s=3, c="#1292db", lw=0, alpha=1, zorder=5)

plt.title("Locations of Drone Strikes")
plt.show()
```
